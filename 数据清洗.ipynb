# 引入pandas plt，对时间进行数据处理

```python

import pandas as pd
import matplotlib.pyplot as plt
file_path = 'data.xlsx'
df=pd.read_excel(file_path, sheet_name='System Data', header=0)
# 按照m/d/y/H:M格式将时间分别储存
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%m/%d/%Y %H:%M')

# 这条代码的作用是将 DataFrame df 中的名为 'DateTime' 的列提升为 DataFrame 的行索引 (Index)，并用新的 DataFrame 替换原有的 df 变量。
# 即把时间作为索引，并且这个很智能，比如df.loc['2023-01-01']就能直接取到2023年1月1日的所有行的数据。
# 例子
df = df.set_index('DateTime')
```

# 可视化2017年的每个月的数据

Figure (图形): 相当于一张画布。它包含所有子图（Axes），以及标题、边框等。

Axes (坐标系/子图): 真正的绘图区域。它包含 X 轴、Y 轴、刻度、标签以及实际的数据图形（如线、点、柱状图等）。一张 Figure 可以包含多个 Axes

```python
import numpy as np
import matplotlib.dates as mdates # 导入日期处理模块
# 显式创建 Figure 和 Axes
fig, ax = plt.subplots(figsize=(15, 6))
#数据准备
x = df.loc['2017-01-01','IE WindA'] # 获取2017年1月1日的IE WindA数据
# 绘制折线图
ax.plot(x.index, x, label='Value over Time')
# A. 设置主刻度定位器为每小时
# Interval=1 表示每隔 1 个单位（小时）设置一个刻度
ax.xaxis.set_major_locator(mdates.HourLocator(interval=1))
# B. 设置刻度标签的格式
# %H:%M 表示以 小时:分钟 的格式显示时间，例如 01:00, 02:00
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))


# 显式创建 Figure 和 Axes
fig, ax = plt.subplots(figsize=(15, 6))
#数据准备
x = df.loc['2017-06-01','IE WindA'] # 获取2017年1月1日的IE WindA数据
# 绘制折线图
ax.plot(x.index, x, label='Value over Time')
# A. 设置主刻度定位器为每小时
# Interval=1 表示每隔 1 个单位（小时）设置一个刻度
ax.xaxis.set_major_locator(mdates.HourLocator(interval=1))
# B. 设置刻度标签的格式
# %H:%M 表示以 小时:分钟 的格式显示时间，例如 01:00, 02:00
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# ==========================================
# 1. 数据准备与预处理
# ==========================================


# 确保没有空值
df = df.dropna()

# 归一化 (Normalization) - 
# 分别对 A 和 B 进行归一化，以便后续反归一化 B 的预测结果
scaler_A = MinMaxScaler(feature_range=(0, 1))
scaler_B = MinMaxScaler(feature_range=(0, 1))


df['IE WindA'] = scaler_A.fit_transform(df[['IE WindA']])
df['IE WindB'] = scaler_B.fit_transform(df[['IE WindB']])



def create_daily_samples(dataframe):
    X = [] # 输入：A场数据
    Y = [] # 标签：B场数据
    
    # 按日期分组
    grouped = dataframe.groupby(dataframe.index.date)
    
    for date, group in grouped:
        if len(group) == 96: # 确保这一天是完整的96个点
            # 获取这一天 A 的归一化数据
            a_data = group['IE WindA'].values
            # 获取这一天 B 的归一化数据
            b_data = group['IE WindB'].values

            X.append(a_data.reshape(96, 1)) 
            Y.append(b_data) # 输出是一维向量 (96,)
            
    return np.array(X), np.array(Y)

X, Y = create_daily_samples(df)

print(f"总样本天数: {len(X)}")
print(f"输入形状: {X.shape} (天数, 96, 1)")
print(f"输出形状: {Y.shape} (天数, 96)")


train_days = 450 # 2016年的天数

X_train = X[:train_days]
Y_train = Y[:train_days]
X_test = X[train_days:]
Y_test = Y[train_days:]

print(f"训练集样本数: {len(X_train)}")
print(f"测试集样本数: {len(X_test)}")

model = Sequential()

# 第一层卷积：提取 A 场出力的局部特征（如爬坡、波峰）
model.add(Conv1D(filters=64, kernel_size=13, activation='relu', input_shape=(96, 1)))
model.add(MaxPooling1D(pool_size=1))

# 第二层卷积
model.add(Conv1D(filters=64, kernel_size=13, activation='relu'))
model.add(MaxPooling1D(pool_size=2))

# 展平层：将卷积提取的特征展平成一维向量
model.add(Flatten())

# 全连接层：建立 A 特征到 B 曲线的映射
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2)) # 防止过拟合

# 输出层：直接输出 96 个点，代表 B 场全天的出力
model.add(Dense(96)) 

# 编译模型
model.compile(optimizer=Adam(learning_rate=0.01), loss='mse') # 回归问题用均方误差 MSE

model.summary()

# ==========================================
# 5. 训练模型
# ==========================================
history = model.fit(
    X_train, Y_train,
    epochs=200,            # 训练轮数，可根据 loss 曲线调整
    batch_size=16,        # 每一批训练的天数
    validation_split=0.1, # 从训练集中分出 10% 验证，监控过拟合
    verbose=1
)

# ==========================================
# 6. 预测与可视化
# ==========================================
# 用测试集 (2017数据) 进行预测
predicted_B_scaled = model.predict(X_train)

# 反归一化：将预测结果还原为真实功率值 MW
predicted_B = scaler_B.inverse_transform(predicted_B_scaled)
real_B = scaler_B.inverse_transform(Y_train)

# 随机抽取一天查看效果
day_idx = 0 # 查看测试集第一天的效果
plt.figure(figsize=(12, 6))
plt.plot(real_B[day_idx], label='Real B Output', color='blue')
plt.plot(predicted_B[day_idx], label='Predicted B Output (from A)', color='red', linestyle='--')
plt.title(f'Wind Farm B Day-Ahead Prediction (Test Sample {day_idx})')
plt.xlabel('Time (15min intervals)')
plt.ylabel('Power Output')
plt.legend()
plt.grid(True)
plt.show()

# 3. 画出训练集的第一天 (index = 0)
day_idx = 0 

plt.figure(figsize=(12, 6))

# 画真实值
plt.plot(X_train[day_idx], label='Real Power (Training Set)', color='blue')

# 画预测值
plt.plot(real_B[day_idx], label='Predicted Power (Training Set)', color='red', linestyle='--')

plt.title(f'Training Set Fit Check - Sample {day_idx}')
plt.xlabel('Time Steps (e.g., 15min intervals)')
plt.ylabel('Power Output (MW)')
plt.legend()
plt.grid(True)
plt.show()

# 绘制 Loss 曲线查看训练过程
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Training Loss')
plt.legend()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error

# ==========================================
# 7. 模型评估指标 (MAPE, RMSE, MAE)
# ==========================================

# --- 方法一：使用 sklearn 库计算 (推荐) ---
# 注意：sklearn 的输出是小数 (例如 0.05 代表 5%)
mape_score = mean_absolute_percentage_error(real_B, predicted_B)

# --- 方法二：使用 Numpy 手动计算 ---
# 防止分母为0的处理：在分母加上一个极小值 (epsilon)
epsilon = 1e-10 
mape_manual = np.mean(np.abs((real_B - predicted_B) / (real_B + epsilon))) * 100

print("-" * 30)
print(f"模型评估结果:")
print(f"MAPE (Sklearn): {mape_score:.2%}")   # 格式化为百分比
print(f"MAPE (Manual):  {mape_manual:.2f}%")
print("-" * 30)

# 补充：通常回归问题也会看 RMSE (均方根误差) 和 MAE (平均绝对误差)
rmse = np.sqrt(mean_squared_error(real_B, predicted_B))
mae = mean_absolute_error(real_B, predicted_B)
print(f"RMSE: {rmse:.4f} MW")
print(f"MAE:  {mae:.4f} MW")
print("-" * 30)

# ==========================================
```
